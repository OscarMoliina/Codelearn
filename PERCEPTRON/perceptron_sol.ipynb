```python
import random

#b = random.random() * 100
w = random.random() * 100

data = {
    1:4,
    4:16,
    2:8,
    20:80
}

def loss(real, predit):
    return sum([(r-p)**2 for r,p in zip(real,predit)])

real = list(data.values())

def neurona(lst:list, w):
    return [w*i for i in lst]

predit = neurona(data.keys(), w)

print(f'W = {w}')
print(loss(real,predit))
```

```python
#w = 100
learning_rate = 0.01
for i in range(10000):
    predit = neurona(data.keys(),w)
    d = neurona(data.keys(), w+learning_rate)
    dif = loss(real, d) - loss(real, predit)
    w -= learning_rate * dif
    print(f'Loss: {round(loss(real, predit),4)} Weigth: {w}')
```
